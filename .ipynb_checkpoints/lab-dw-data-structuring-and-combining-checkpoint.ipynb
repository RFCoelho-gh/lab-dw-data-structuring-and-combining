```python
# Your code goes here

import pandas as pd

url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
```

```python
list_of_df = [df1, df2, df3]
```

```python
for d in list_of_df:
    d.columns = [col.lower().replace(" ","_") for col in d.columns]
    if "st" in d.columns:
        d.rename(columns= {'st': 'state'}, inplace=True)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
insurance_df = pd.concat([df1, df2, df3], axis=0)
```

```python
insurance_df
```

```python
insurance_df.dropna(subset=['customer'], inplace=True)
```

```python
insurance_df.isna().sum()
```

```python
insurance_df["customer"].duplicated().sum()
```

```python
insurance_df.drop_duplicates(subset=["customer"], inplace=True)
```

```python
insurance_df["customer"].duplicated().sum()
```

```python
insurance_df.isna().sum()
```

```python
insurance_df.gender.fillna("NA", inplace=True)
```

```python
insurance_df.isna().sum()
```

```python
insurance_df["gender"].value_counts()
```

```python
insurance_df['gender'] = insurance_df['gender'].map({"M": 'Male', 'F': 'Female', 'Femal' : 'Female', 'female' : 'Female', 'Male' : 'Male'})
```

```python
insurance_df["gender"].value_counts()
```

```python
insurance_df["state"].value_counts()
```

```python
def state_convertor(x):
    if x == "WA":
        return "Washington"
    elif x == "AZ":
        return "Arizona"
    elif x == "Cali":
        return "California"
    else:
        return x
```

```python
insurance_df["state"] = insurance_df["state"].apply(state_convertor)
```

```python
insurance_df["state"].value_counts()
```

```python
insurance_df["education"].value_counts()
```

```python
def bacherlor_slicer(x):
    if x == "Bachelors":
        return "Bachelor"
    else:
        return x
```

```python
insurance_df["education"] = insurance_df["education"].apply(bacherlor_slicer)
```

```python
insurance_df["education"].value_counts()
```

```python
insurance_df["customer_lifetime_value"].nunique()
```

```python
insurance_df["customer_lifetime_value"].value_counts().sum()
```

```python
def lifetime_spliter(x):
    if type(x) == str:
        temp = float(x.split("%")[0])
        return temp
    else:
        return x
```

```python
insurance_df["customer_lifetime_value"] = insurance_df["customer_lifetime_value"].apply(lifetime_spliter)
```

```python
insurance_df["customer_lifetime_value"].value_counts().sum()
```

```python
insurance_df["vehicle_class"].value_counts()
```

```python
insurance_df["vehicle_class"].value_counts().sum()
```

```python
def luxury_vehicle(x):
    luxury_cars = ["Sports Car", "Luxury SUV", "Luxury Car"]
    if x in luxury_cars:
        return "Luxury"
    else:
        return x
```

```python
insurance_df["vehicle_class"] = insurance_df["vehicle_class"].apply(luxury_vehicle)
```

```python
insurance_df["vehicle_class"].value_counts()
```

```python
def split_complaints(x):
    if type(x) == str:
        temp = x.split("/")[1]
        if temp.isdigit():
            return int(temp)
        else:
            return x
    else:
        return x
```

```python
insurance_df["number_of_open_complaints"] = insurance_df["number_of_open_complaints"].apply(split_complaints)
```

```python
insurance_df["number_of_open_complaints"].value_counts()
```

```python
insurance_df["number_of_open_complaints"].value_counts().sum()
```

```python
insurance_df.head(10)
```

```python
insurance_df.isna().sum()
```

```python
insurance_df.gender.fillna("NA", inplace=True)
```

```python
insurance_df.isna().sum()
```

```python
insurance_df.customer_lifetime_value.fillna(insurance_df.customer_lifetime_value.mean(), inplace=True)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Your code goes here

url4 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"

marketing_df = pd.read_csv(url4)
```

```python
marketing_df
```

```python
marketing_df.customer_lifetime_value.sum()
```

```python
def rounder(x):
    return round(x,2)
```

```python
z = 2.5768547534583495
rounder(z)
```

```python
marketing_df["customer_lifetime_value"] = marketing_df["customer_lifetime_value"].apply(rounder)
```

```python
marketing_df.head()
```

```python
marketing_df.columns
```

```python
marketing_df["sales_channel"].value_counts()
```

```python
#marketing_df["customer_lifetime_value"] = marketing_df["customer_lifetime_value"].apply(rounder)
# after transformation, the rounding is LOST!
```

```python
mkt_pivot = marketing_df.pivot_table(index="sales_channel",columns="policy_type", values="customer_lifetime_value")
```

```python
mkt_pivot
```

```python
auto_list = ["Corporate Auto", "Personal Auto", "Special Auto"]
for i in auto_list:
    mkt_pivot[i] = mkt_pivot[i].apply(rounder)
```

```python
mkt_pivot
```

- Corporate Auto has a homogenous success across all types of sales channels. Branch is marginally higher than the others.

- Personal Auto follows a similar tendency, but has a more accentued success in Call Center.

- Special Auto are the most successful through agents, with a clear difference between the top channel and the others. We can assume that our customers with the Special Auto insurance are more likely to sign through in-person agents.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
marketing_df.groupby(["gender", "education"])["customer_lifetime_value"].mean()
```

- With female customers, the average lifetime value is negatively correlated with an increase of the education level. Our most valuable group are female customers with high school degrees or lower.

- With male customers, the average lifetime value rises with their education level, with the exeception of doctors. Our most valuable groups are male customers with college, high school or master degreeds.
